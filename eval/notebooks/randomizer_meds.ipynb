# Code to filter list for evaluation

```python
import pandas as pd
```

```python
df = pd.read_csv("medlist.csv", sep=';', encoding="utf-8")
```

Apply 'filter' to the list

```python

filtered_df = df[
    ~df['Bezeichnung'].str.contains(r'injektion|infusion|salbe|gel|creme|pflaster|trockenstechampulle|pricktest|augentropfen|nasenspray|konzentrat', case=False, na=False) &  
    ~df['Zulassungsnummer'].str.startswith('EU', na=False) &                            # Exclude rows where 'Zulassungsnummer' starts with "EU"
    ~(df['Einstellung In-Verkehr-Bringen'].str.lower() == 'ja') &
    (df['Suchtgift'] != 1)  # Exclude rows where 'Suchtgift' equals 1                   # No anesthetics or narcotics  
]
```

Result - filtered list

```python
filtered_df
```

```python
# Select 10 random rows from the filtered DataFrame
random_sample = filtered_df.sample(n=10)
```
